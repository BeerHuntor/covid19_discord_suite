```python
import os
import json
from pandas.io.json import json_normalize
import pandas as pd
```

```python
pwd = os.getcwd()
filepath = (pwd + '/data_sets/')
```

```python
with open(filepath + 'website_data.json') as f:
    d = json.load(f)
```

```python
data = json_normalize(d['data'])
data.head()
```

```python
list(data.columns)
```

## DECIDED TO KEEP THE DEMOGRAPHIC DEATH STATS SEPERATE FROM THE REST 

**This I feel will be easier to work with in the long run due to the nature of only having a 'daily' figure**

```python
phase_one_data = data.copy()
female_death_data = pd.json_normalize(d['data'], record_path=['femaleDeaths'],
meta=['date', 'name', 'code', 'totalCases', 'totalCaseRate', 'hospitalisations','ventBeds', 'deaths', 'deathRate','hospitalCases'], errors='ignore')

female_death_data.rename(columns={'age' : 'd_age', 'rate' : 'f_d_a_rate', 'value' : 'f_total_deaths_by_age'}, inplace=True)

female_death_data.drop(columns=['code'], inplace=True)
female_death_data
```

```python
male_death_data = pd.json_normalize(d['data'], record_path=['maleDeaths'],
meta=['date'], errors='ignore')

male_death_data.rename(columns={'age' : 'd_age', 'rate' : 'm_d_a_rate', 'value' : 'm_total_deaths_to_date'}, inplace=True)
male_death_data.drop(columns=['d_age'], inplace=True)
male_death_data
```

```python
female_death_data['m_d_a_rate'] = male_death_data['m_d_a_rate']
female_death_data['m_total_deaths_to_date'] = male_death_data['m_total_deaths_to_date']
demographic_death_data = female_death_data.copy()

###REARRANGE COLUMNS
columnTitles = ['date', ]
demographic_death_data
```
