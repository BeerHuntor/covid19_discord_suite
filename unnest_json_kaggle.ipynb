```python
import os
import json
from pandas.io.json import json_normalize
import pandas as pd
```

```python
pwd = os.getcwd()
filepath = (pwd + '/data_sets/')
```

```python
with open(filepath + 'demographic_sex_death_data.json') as f:
    d = json.load(f)
```

```python
data = json_normalize(d['data'])
data.head()
```

```python
list(data.columns)
```

## DECIDED TO KEEP THE DEMOGRAPHIC DEATH STATS SEPERATE FROM THE REST 

**This I feel will be easier to work with in the long run due to the nature of only having a 'daily' cumalative figure**

```python
phase_one_data = data.copy()
phase_two_data = data.copy()
phase_one_data = json_normalize(d['data'], record_path=['female_28_days'],
meta=['date', 'area'])
phase_two_data = json_normalize(d['data'], record_path=['male_deaths_28_days'])
# phase_one_data = json_normalize(d['data'], record_path=['female_28_days'],
# meta=['date', 'area', 'male_deaths_28_days'], errors='ignore')

#phase_one_data.rename(columns={'age' : 'f_age', 'rate' : 'f_death_rate', 'value' : 'f_death_num'}, inplace=True)
phase_one_data.rename(columns={'rate' : 'f_death_rate_per_100k', 'value' : 'f_total_deaths'}, inplace=True)
phase_one_data
```

```python
phase_two_data.rename(columns={'rate' : 'm_death_rate_per_100k', 'value' : 'm_total_deaths'}, inplace=True)
phase_two_data
```

```python
final_phase_data = phase_one_data.join(phase_two_data.set_index('age'), on='age')
final_phase_data = (final_phase_data.reindex(['date', 'area','f_death_rate_per_100k', 'f_total_deaths', 'age', 'm_death_rate_per_100k', 'm_total_deaths'], axis=1))
final_phase_data
```
